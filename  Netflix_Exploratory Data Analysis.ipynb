```python
import numpy as np
import pandas as pd
import plotly.express as px
from textblob import TextBlob


df=pd.read_csv('netflix_titles.csv')


df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
x=df.groupby(['rating']).size().reset_index(name='counts')
print(x)
```

```python
pieChart = px.pie(x , values='counts' , names='rating')
pieChart.show()
```

```python
df['director']=df['director'].fillna('Director not specified')
df.head()
```

```python
director_list=pd.DataFrame()


director_list= df['director'].str.split(',' , expand=True).stack()

print(director_list)
```

```python
director_list = director_list.to_frame()
print(director_list)
```

```python
director_list.columns=['Directors']
print(director_list)
```

```python
director=director_list.groupby(['Directors']).size().reset_index(name='Total Count')
print(director)
```

```python
director=director[director.Directors != 'Director not specified']
print(director)
```

```python
director=director.sort_values(by=['Total Count'] , ascending=False)
print(director)
```

```python
top5Directors=director.head()
print(top5Directors)
```

```python
barChart = px.bar(top5Directors , x='Total Count' , y='Directors' )
barChart.show()
```
